<h2> CHUYỂN DỮ LIỆU CỘT THỂ LOẠI THÀNH DẠNG SỐ <h2>

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("genre.csv")
df.head()
```

```python
df['Lãng Mạn'] = df['genre'].apply(lambda x: 1 if 'Lãng Mạn' in x else 0)

df['Tình Tiết'] = df['genre'].apply(lambda x: 1 if 'Tình Tiết' in x else 0)

df['Thanh Xuân'] = df['genre'].apply(lambda x: 1 if 'Thanh Xuân' in x else 0)

df['Thành Thị'] = df['genre'].apply(lambda x: 1 if 'Thành Thị' in x else 0)

df['Viễn Tưởng'] = df['genre'].apply(lambda x: 1 if 'Viễn Tưởng' in x else 0)

df['Phim Hài'] = df['genre'].apply(lambda x: 1 if 'Phim Hài' in x else 0)

df['Bí Ẩn'] = df['genre'].apply(lambda x: 1 if 'Bí Ẩn' in x else 0)

df['Hành Động'] = df['genre'].apply(lambda x: 1 if 'Hành Động' in x else 0)

df['Cổ Trang'] = df['genre'].apply(lambda x: 1 if 'Cổ Trang' in x else 0)

df['Tình Bạn'] = df['genre'].apply(lambda x: 1 if 'Tình Bạn' in x else 0)

df['Phiêu Lưu'] = df['genre'].apply(lambda x: 1 if 'Phiêu Lưu' in x else 0)

df['Tội Phạm'] = df['genre'].apply(lambda x: 1 if 'Tội Phạm' in x else 0)

df['Tình Yêu Ngọt Ngào'] = df['genre'].apply(lambda x: 1 if 'Tình Yêu Ngọt Ngào' in x else 0)

df['Gia Đình'] = df['genre'].apply(lambda x: 1 if 'Gia Đình' in x else 0)

df['Nhiệt Huyết'] = df['genre'].apply(lambda x: 1 if 'Nhiệt Huyết' in x else 0)

df['Kinh Dị'] = df['genre'].apply(lambda x: 1 if 'Kinh Dị' in x else 0)

df['Tiểu Thuyết Chuyển Thể'] = df['genre'].apply(lambda x: 1 if 'Tiểu Thuyết Chuyển Thể' in x else 0)

df['Other'] = df['genre'].apply(lambda x: 1 if any(i for i in x.split(', ') if i not in ['Lãng Mạn','Tình Tiết','Thanh Xuân','Thành Thị','Viễn Tưởng','Phim Hài','Bí Ẩn','Hành Động','Cổ Trang','Tình Bạn','Phiêu Lưu','Tội Phạm','Tình Yêu Ngọt Ngào',	'Gia Đình' ,'Nhiệt Huyết','Kinh Dị','Tiểu Thuyết Chuyển Thể' ]) else 0)
```

```python
df.to_csv("genre1.csv", index = False)
df.head()
```
